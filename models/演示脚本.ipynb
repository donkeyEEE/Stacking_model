```python
# 对ECFP参数进行灵敏度分析
len_lis = [512, 1024, 2048, 4096]
e_lis = [2, 4, 6]
```

```python
dic_RF={}
for l in len_lis:
    dic_RF[l] =[]
    for r in e_lis:
        # 定义参数类
        args = Args(train, test, AFP=False, RF=True, MPNN=False,
                    SVR=False, save_r=False, plot=True,GAT=True,
                    AD_FP=False)
        # 实例化集成模型
        eb = ensemble_models()
        print("[{},{}]".format(l,r))
        args.ECFP_Params=[l,r]
        eb.Lay1_models_train(args ,save=False)
        out_RF = metric_r(eb.L1_test_df['true'], eb.L1_test_df['RF'])
        print(out_RF)
        dic_RF[l].append(out_RF)
```

```python
def la(x):  # EPA标签
    x = 10 ** x
    if x < 500:
        return 0
    else:
        return 1
    return None
metric_c(eb.L1_test_df['true'].map(la), eb.L1_test_df['RF'].map(la))
```
